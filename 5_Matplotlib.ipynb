```python
months=[1,2,3,4]
sales=[100,150,1200,2000]
plt.plot(months,sales,color="blue",linestyle="--",linewidth=1,marker="o",label="2026 sales data")
plt.xlabel("months")
plt.ylabel("sales per month")
plt.title("monthly sales per month")
plt.legend(loc="lower right",fontsize=6)
plt.grid(color="gray",linestyle=":",linewidth=1)
plt.show()
```

```python
months=[1,2,3,4]
sales=[100,150,1200,2000]
plt.plot(months,sales,color="blue",linestyle="--",linewidth=1,marker="o",label="2026 sales data")
plt.xlabel("months")
plt.ylabel("sales per month")
plt.title("monthly sales per month")
plt.legend(loc="lower right",fontsize=6)
plt.grid(color="gray",linestyle=":",linewidth=1)
plt.xlim(1,4)
plt.ylim(0,2000)
plt.xticks([1,2,3,4],['M1','M2','M3','M4'])
plt.show()
```

### Scatterplot

```python
x=[1,2,3,4,5,6]
y=[2,3,4,5,6,7]

plt.scatter(x,y,color="blue",marker="o")
```

```python
hours_studies=[1,2,3,4,5,6,7,8]
exam_scores=[50,55,60,65,70,75,80,85]
plt.scatter(hours_studies,exam_scores,color="green",marker="o",label="student data")
plt.xlabel("hour studied")
plt.ylabel("exam score")
plt.title("relationship between study time and exam score")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.scatter([1,2,3],[50,55,60], color="blue",label="class1")
plt.scatter([1,2,3],[45,50,52],color="orange",label="class2")
plt.xlabel("hour studies")
plt.ylabel("exam score")
plt.title("comparison of two class")
plt.legend()
plt.grid(True)
plt.show()
```

### Pie chart

```python
labels=['A','B','C','D']
sizes=[30,20,40,10]
colors=['pink','blue','red','grey']
explode=[0.2,0,0,0]

plt.pie(sizes,labels=labels,explode=explode,colors=colors,autopct="%1.1f%%",shadow=True)
```

```python
regions=['north','south','east','west']
revenue=[3000,2000,1500,1000]
explodes=[0.2,0,0,0]
plt.pie(revenue,labels=regions,autopct='%1.1f%%',colors=["green","blue","orange","brown"],explode=explodes)
plt.title("Revenue contribution per region")
plt.show()
```

### Subplot

```python
x=[1,2,3,4,5]
y1=[1,4,9,16,25]
y2=[1,2,3,4,5]

plt.figure(figsize=(9,5))

plt.subplot(2,2,1)
plt.plot(x,y1,color="green")
plt.title("plot1")

plt.subplot(2,2,2)
plt.plot(y1,x,color="green")
plt.title("plot1")

plt.subplot(2,2,3)
plt.plot(x,y2,color="green")
plt.title("plot1")

plt.subplot(2,2,4)
plt.plot(x,y2,color="green")
plt.title("plot1")
```

```python
x=[1,2,3,4]
y=[10,15,20,25]

plt.figure(figsize=(9,5))

plt.subplot(1,2,1)
plt.plot(x,y)
plt.title("line chart")

plt.subplot(1,2,2)
plt.bar(x,y)
plt.title("bar chart")

plt.show()
```

```python
fig,ax=plt.subplots(1,2,figsize=(10,5))

x=[1,2,3,4]
y=[10,20,15,25]

ax[0].plot(x,y,color="green")
ax[0].set_title("line plot")

ax[1].bar(x,y,color="red")
ax[1].set_title("bar chart")

plt.tight_layout()
plt.show()
```
