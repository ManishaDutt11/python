***decorator***

```python
# adding functionality on the function without changing the main funcation code

def my_decorator(func):
    def wrapper():
        print("Something is happening before the function is called")
        func()
        print("Something is happening after the function is called")
    return wrapper

@my_decorator
def say_hello():
    print("Hello")

say_hello()
```

***operator overloading***

```python
class vector:
    def __init__(self,i,j):
        self.i=i
        self.j=j

    def __add__(self,other):
        return vector(self.i+other.i,self.j+other.j)

    def __str__(self):
        return f"Vector({self.i},{self.j})"

v1=vector(1,2)
v2=vector(3,4)

v3=v1+v2
print(v3)
```

***encapsulation***

```python
class Person:
    def __init__(self, name, age):
        self.__name = name  # private variable
        self.__age = age    # private variable

    def get_name(self):
        return self.__name

    def set_name(self, name):
        self.__name = name

    def get_age(self):
        return self.__age

    def set_age(self, age):
        if age > 0:
            self.__age = age
        else:
            print("Age cannot be negative")

person = Person("Manisha", 22)
print(person.get_name())  
print(person.get_age())   
person.set_age(35)
print(person.get_age())   
person.set_age(-5)        
```

***abstraction***

```python
from abc import ABC, abstractmethod

class Vehicle(ABC):
    @abstractmethod
    def start(self):
        pass

class Car(Vehicle):
    def start(self):
        print("car start with key")

class Bike(Vehicle):
    def start(self):
        print("bike start with button")


car=Car()
bike=Bike()

car.start()
bike.start()
```

```python
try:
    num=int(input("enter a number:"))
    result=10/num
    print(result)

except ZeroDivisionError:
    print("you can't divide number by zero")

except ValueError:
    print("you can't divide bt string")

    
```

```python
# nested

try:
    num1=int(input("enter a first number:"))
    num2=int(input("enter a second number:"))
    try:
        print(num1/num2)
    except ZeroDivisionError:
        print("you cannot divide by zero")
except ValueError:
    print("you cannot divide by string")
    
```

```python
# password character checking

def password(p):
    if len(p)<8:
        raise Exception("error:password must be >=8")
    print("password is strong")

try:
    passwords=input("enter a password:")
    password(passwords)
except Exception as e:
    print(e)
```

```python
try:
    num=int(input("enter a number:"))
    result=10/num
    print(result)

except ZeroDivisionError as e:
    print(e)
```

```python
# else
try:
    num = int(input("Enter a number: "))
    result = 10 / num
except ValueError:
    print("Not a valid number.")
except ZeroDivisionError:
    print("You can't divide the number by zero.")
except Exception as x:
    print(x)
else:
    print(f"The result is {result}")
```

```python
# finallt

try:
    num=int(input("enter a number:"))
    result=10/num
    print(result)

except ZeroDivisionError:
    print("you can't divide number by zero")

except ValueError:
    print("you can't divide bt string")

finally:
    print("that was the concept of exception handling")
```
