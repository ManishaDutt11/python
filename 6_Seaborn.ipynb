```python
import seaborn as sns
```

```python
# basic plotting

tips=sns.load_dataset('tips')
tips
```

```python
# scatterplot

import matplotlib.pyplot as plt
```

```python
sns.scatterplot(x="total_bill",y="tip",data=tips)
plt.title("Scatter plot of total bill vs tip")
plt.grid(True)
plt.show()
```

```python
# line plot

sns.lineplot(x="size",y='total_bill',data=tips)
plt.title("line plot of total bill by size")
plt.show()
```

```python
# categorical plots

sns.barplot(x='day',y='total_bill',data=tips)
plt.title("bar plot of total bill by day")
plt.show()
```

```python
# BOX PLOT

sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
# violin plot

sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
# histogram

sns.histplot(tips['total_bill'],bins=10,kde=True)
```

```python
# Kde plot

sns.kdeplot(tips['total_bill'],fill=True)
```

```python
# pairplot

sns.pairplot(tips)
```

```python
# heatmap

corr=tips[['total_bill','tip','size']].corr()
corr
```

```python
sns.heatmap(corr,annot=True,cmap='coolwarm')
```

```python
import pandas as pd
```

```python
df=pd.read_csv("sales_data_sample (2).csv",encoding="latin1")
```

```python
df
```

```python
plt.figure(figsize=(10,6))

sns.barplot(x='CITY',y='SALES',data=df,estimator=sum)
plt.title("sales throughout the city")
plt.xlabel("city")
plt.xticks(rotation=90)
plt.ylabel("total sales")
plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.barplot(x='COUNTRY',y='SALES',data=df,estimator=sum)
plt.title("sales throughout the country")
plt.xlabel("country")
plt.xticks(rotation=90)
plt.ylabel("total sales")
plt.show()
```
