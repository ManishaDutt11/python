```python

# The outer loop decides how many rows you stamp.

# The inner loop decides how many stars in each row.

# end=" " tells you to stamp stars in the same row.

# The final print() means: Now move to the next line and start stamping again.
```

```python
n=10

for i in range(1,n):
    for j in range(1,n):
        print("*",end=" ")
    print()
```

```python
n=10

for i in range(1,n+1):
    for j in range(1,n+1):
        if i==1 or i==n or j==1 or j==n:
            print("*",end=" ")
        else:
            print(" ",end=" ")
    print()
```
